```python
import math
import pandas as pd
import numpy as np

from pathlib import Path

import plotly
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

pio.renderers.default = 'notebook_connected'
```

```python
DATA_DIR = Path.cwd().parent / "data"
CSV_WHITE_NOISE = DATA_DIR / "spectra_white_noise.csv"
CSV_PINK_NOISE = DATA_DIR / "spectra_pink_noise.csv"
CSV_MEANDER_NOISE = DATA_DIR / "spectra_meander_noise.csv"
CSV_BOAT_NOISE = DATA_DIR / "spectra_boat_noise.csv"
CSV_CHANNEL_TO_FREQUENCY = DATA_DIR / "channel_frequency.csv"
```

```python
channel_to_frequency_df = pd.read_csv(CSV_CHANNEL_TO_FREQUENCY)
channel_to_frequency_df.columns = ["channel", "frequency"]

white_noise_df = pd.read_csv(CSV_WHITE_NOISE)
white_noise_df["channel"] = channel_to_frequency_df["frequency"]
white_noise_df.columns = ["frequency", "amplitude"] # Hz, dB

pink_noise_df = pd.read_csv(CSV_PINK_NOISE)
pink_noise_df["channel"] = channel_to_frequency_df["frequency"]
pink_noise_df.columns = ["frequency", "amplitude"] # Hz, dB

meander_noise_df = pd.read_csv(CSV_MEANDER_NOISE)
meander_noise_df["channel"] = channel_to_frequency_df["frequency"]
meander_noise_df.columns = ["frequency", "amplitude"] # Hz, dB

boat_noise_df = pd.read_csv(CSV_BOAT_NOISE)
boat_noise_df["channel"] = channel_to_frequency_df["frequency"]
boat_noise_df.columns = ["frequency", "amplitude"] # Hz, dB
```

```python
# conversion to spectral level
def to_spectral_level(row):
    frequency: float = row["frequency"]
    sound_level: float = row["amplitude"]
    return sound_level - 20 * math.log10(math.sqrt(0.23 * frequency))
    
white_noise_df["amplitude"] = white_noise_df.apply(to_spectral_level, axis=1)
pink_noise_df["amplitude"] = pink_noise_df.apply(to_spectral_level, axis=1)
meander_noise_df["amplitude"] = meander_noise_df.apply(to_spectral_level, axis=1)
boat_noise_df["amplitude"] = boat_noise_df.apply(to_spectral_level, axis=1)
```

```python
X_WHITE_NOISE = white_noise_df["frequency"]
Y_WHITE_NOISE = white_noise_df["amplitude"]

X_PINK_NOISE = pink_noise_df["frequency"]
Y_PINK_NOISE = pink_noise_df["amplitude"]

X_MEANDER_NOISE = meander_noise_df["frequency"]
Y_MEANDER_NOISE = meander_noise_df["amplitude"]

X_BOAT_NOISE = boat_noise_df["frequency"]
Y_BOAT_NOISE = boat_noise_df["amplitude"]
```

```python
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("White noise", "Pink noise", "Meander noise", "Boat noise"),
)

fig.add_trace(go.Scatter(x=X_WHITE_NOISE, y=Y_WHITE_NOISE, mode="lines+markers", marker=dict(size=8)), row=1, col=1)
fig.update_xaxes(title_text=r"$Frequency, Hz$", row=1, col=1, type="log")
fig.update_yaxes(title_text=r"$\alpha, dB$", row=1, col=1)

fig.add_trace(go.Scatter(x=X_PINK_NOISE, y=Y_PINK_NOISE, mode="lines+markers", marker=dict(size=8)), row=1, col=2)
fig.update_xaxes(title_text=r"$Frequency, Hz$", row=1, col=2, type="log")
fig.update_yaxes(title_text=r"$\alpha, dB$", row=1, col=2)

fig.add_trace(go.Scatter(x=X_MEANDER_NOISE, y=Y_MEANDER_NOISE, mode="lines+markers", marker=dict(size=8)), row=2, col=1)
fig.update_xaxes(title_text=r"$Frequency, Hz$", row=2, col=1, type="log")
fig.update_yaxes(title_text=r"$\alpha, dB$", row=2, col=1)

fig.add_trace(go.Scatter(x=X_BOAT_NOISE, y=Y_BOAT_NOISE, mode="lines+markers", marker=dict(size=8)), row=2, col=2)
fig.update_xaxes(title_text=r"$Frequency, Hz$", row=2, col=2, type="log")
fig.update_yaxes(title_text=r"$\alpha, dB$", row=2, col=2)

fig.update_layout(
    title_text="Noise spectras",
    height=1200,
    showlegend=False,
    template="plotly_dark",
)

fig.show()
```
